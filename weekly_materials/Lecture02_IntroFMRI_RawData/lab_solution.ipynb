1- [2pts] Load the data from the last time, and then transpose the matrix to get the order: Time, Z, Y, X

```python
### STUDENT ANSWER
# import relevant packages
import numpy as np
import matplotlib.pyplot as plt  # for visualization

# Set defaults for matplotlib plotting in the notebook
%matplotlib inline

# load data and transpose:
fname = '/home/jovyan/data/sub01_categories1_1.npy'
data = np.load(fname) 
data = data.astype('float32')
data = data.T

print("data shape :", data.shape)
```

2- [1pt] Create a transverse slice using X=45 and Y=45.

```python
### STUDENT ANSWER
transverse = data[:,:,45,45]
```

3- [1pt] Use np.min to make the minimum value of each column 0 as we did in class.

```python
### STUDENT ANSWER
transverse_min0 = transverse - np.min(transverse)
```

4- [4pts] Use np.max to make the maximum value of each column 1. Think of what this means. How can you do it? Explain your reasoning using comments (to write comments in python, start a line with #).

```python
### STUDENT ANSWER
transverse_max1 = transverse_min0 / np.max(transverse_min0)
# the line above makes the previous array which had a minimum of 0 for all columns, now ALSO have a maximum of 1.
# If I only do transverse_max1 = transverse / np.max(transverse)
# then the maximum will be 1, but the minimum is not garanteed to be 0.
```
